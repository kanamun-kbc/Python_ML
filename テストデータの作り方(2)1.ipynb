```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
import sympy
import math
```

## (1)の回答 サンプルデータの作成

```python
 # xの値
x = np.linspace(1.0, 5.5, 100)
 # yの真の式
y = 2.1*x - 0.9
 # ノイズ付の式
np.random.seed(42)
y_noise = 2.1*x - 0.9 + np.random.uniform(-0.5, 0.5, 100)
```

## (2)の回答 学習データへの分類と単回帰分析

```python
# 学習データとテストデータに分ける
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y_noise, test_size=0.2, random_state=42)

# 単回帰分析
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.reshape(-1,1), y_train.reshape(-1,1))
print(f'学習データから、y = {model.coef_[0,0]}*x + {model.intercept_[0]} と近似できる')
```

## (3)の回答 真の値と予測値との平均絶対値誤差

```python
 # 予測結果
y_pred = model.predict(x_test.reshape(-1,1))
 # 真の値
y_true = 2.1*x_test.reshape(-1,1) - 0.9
 # 平均絶対値誤差
mae = np.mean(np.abs(y_true - y_pred))
print(f'平均絶対値誤差の値は {mae} である。')
```
