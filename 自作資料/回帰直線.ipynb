```python
import sympy

# 微分して極値点求めた
a = sympy.Symbol('a')
b = sympy.Symbol('b')

f =(3.1-(a * 1.0 + b))**2  + (5.1-(a * 2.0 + b))**2+ (6.8-(a * 3.0 + b))**2+ (8.9-(a * 4.0 + b))**2 + (11.5-(a * 5.0 + b))**2 + (13.8-(a * 6.0 + b))**2

da = sympy.diff(f, a)
db = sympy.diff(f, b)

print(da)
print(db)

list = sympy.solve([da, db])
print(list)
```

```python
# 最小二乗法で回帰直線を求める(a, b を求める)

import numpy as np
import pandas as pd

df_shokubutsu = pd.DataFrame([[1, 3.1], [2, 5.1], [3, 6.8], [4, 8.9], [5, 11.5], [6, 13.8]])
df_shokubutsu.columns = ['X','y']
print(df_shokubutsu)

def linear_regression(df):
    x = df['X'].to_numpy()
    y = df['y'].to_numpy()
    x_mean = np.mean(x)
    y_mean = np.mean(y)

#    NSxy = np.sum((x - x_mean)*(y - y_mean))  # Sxy を N 倍した値
#    NSxx = np.sum((x - x_mean)**2)            # Sxx を N 倍した値
    Sxy = np.cov([x,y],ddof=0)[0,1]
    Sxx = np.var(x)
    a = Sxy / Sxx
    b = y_mean - a * x_mean
    return a, b

a_ans, b_ans = linear_regression(df_shokubutsu)
print(f'a = {a_ans}, b ={b_ans}')
```
