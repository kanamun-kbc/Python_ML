回帰用

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import japanize_matplotlib

X, y, coef = make_regression(random_state=12, n_samples=20, n_features=1,
                             noise=5.0, bias=0.0, coef=True)
print(f'X ={X}')
print(f"Y ={y}")
print(f"coef ={coef}")

plt.title("make_regressionによるテストデータ")
plt.xlabel("X")
plt.ylabel("y")
plt.scatter(X,y)
plt.grid()
plt.show()
```

平行移動

```python
from sklearn.datasets import make_blobs,make_moons
import matplotlib.pyplot as plt
import japanize_matplotlib

X, y = make_moons(n_samples=100, noise=0.05, random_state=42)
X_moved = X + 1.2
# 2つの説明変数それぞれに1.2を足している
plt.scatter(X[:,0:1],X[:,1:2], c = y, alpha=1.0)
plt.scatter(X_moved[:,0:1],X_moved[:,1:2], c = y, alpha=0.4)
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid()
plt.show()
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import japanize_matplotlib

X, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=42, cluster_std=0.8)
X_moved = X + 2.5

plt.scatter(X[:,0:1],X[:,1:2], c = y, alpha=0.9)
plt.scatter(X_moved[:,0:1],X_moved[:,1:2], c = y, alpha=0.4)
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid()
plt.show()
```

色変え

```python
import numpy as np
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import japanize_matplotlib

X, y = make_moons(n_samples=100, noise=0.05, random_state=42)
#print(X)
#print(y)


color = np.array([("r" if i==1 else "g") for i in y])
#print(color) を実行してみよう。何が表示される？
plt.scatter(X[:,0:1],X[:,1:2], c = color, alpha = 0.9)
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()

```

CSV保存方法

```python
import numpy as np
from sklearn.datasets import make_blobs,make_moons

X, y = make_moons(n_samples=100, noise=0.05, random_state=42)
#print(X)
#print(y)

# concatenateは配列の結合(2年前期第５回)
# Xは100行2列
# yは1行100列なので100行1列にしたい
data001 = np.concatenate([X, y.reshape(-1,1)], 1)
print(data001)
# １列にデータが全て並べられてしまう
# np.savetxt("20230926test01.csv",data001)
# 指定すれば３列に分けて並べられる
np.savetxt("20230926test02.csv",data001, header='X1,X2,y', comments="", delimiter=",")
```
