```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
import random
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor

df = pd.read_csv("/content/drive/MyDrive/curve.csv")
print(df)
x = df["x"].to_numpy().reshape(-1,1)
y = df["y"].to_numpy()

# model = RandomForestRegressor(max_depth=3, random_state=0)
# model = KNeighborsRegressor(n_neighbors=2)
# model = SVR()
# model = MLPRegressor(hidden_layer_sizes=(300,300),random_state=1, max_iter=900)
model = MLPRegressor(hidden_layer_sizes=(300,300,300),random_state=1, max_iter=900)

# 学習させる
model.fit(x,y)
#model.fit(x[0:80],y[0:80])

# xを代入した値を予測
y_pred = model.predict(x)
plt.plot(x, y, c="k", label='教師データ')
plt.scatter(x, y_pred, c="r", label='教師データをモデルが近似した結果')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(fontsize = 15)
plt.grid()
plt.show()

print(f'x = 10 のときの y の予測値は {model.predict([[10.0]])}')
```
