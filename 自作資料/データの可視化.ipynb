```python
import pandas as pd                    #　以下略
import numpy as np                     #　以下略
import matplotlib.pyplot as plt        #　以下略
import japanize_matplotlib             #　以下略
import seaborn as sns                  #　以下略

df = pd.read_csv("/content/drive/MyDrive/iris.csv",sep=",")   #　以下略

# relplot(1)　スタイルの設定

sns.set(style='dark')
#darkgrid：背景に色とグリッドがつく。　whitegrid'：白背景にグリッドが付く。
#dark：背景に色がつく。            　white：元の白背景
sns.relplot(data=df,x='sepal_length', y='petal_length',color='tomato')
#color = '色'で点の色を変えられる。
plt.show()
```

```python
import pandas as pd                    #　以下略
import numpy as np                     #　以下略
import matplotlib.pyplot as plt        #　以下略
import japanize_matplotlib             #　以下略
import seaborn as sns                  #　以下略

df = pd.read_csv("/content/drive/MyDrive/iris.csv",sep=",")   #　以下略

# relplot(2)　カテゴリー別に色を分ける。マーカーを分ける。グラフを分ける。

sns.set(style='darkgrid')
# hueで種ごとに色分け
sns.relplot(data=df,x='sepal_length', y='petal_length', hue='species')
# styleでマーカーの形を分ける
sns.relplot(data=df,x='sepal_length', y='petal_length', hue='species', style='species')
# colでグラフを分ける
sns.relplot(data=df,x='sepal_length', y='petal_length', hue='species', col='species')
plt.show()
```

```python
import pandas as pd                    #　以下略
import numpy as np                     #　以下略
import matplotlib.pyplot as plt        #　以下略
import japanize_matplotlib             #　以下略
import seaborn as sns                  #　以下略

df = pd.read_csv("/content/drive/MyDrive/iris.csv",sep=",")   #　以下略

sns.set(style='darkgrid')
sns.catplot(data=df,x='species', y='sepal_length', kind ='box',palette='dark')
#'box'：箱ひげ図（box plot）を作成します。カテゴリごとの分布の要約統計量を示します。
#'violin'：バイオリンプロット（violin plot）を作成します。カテゴリごとの分布をカーネル密度推定に基づいて表示します。
#'bar'：棒グラフ（bar plot）を作成します。カテゴリごとの集計値を棒状に表示します。
plt.show()
```
