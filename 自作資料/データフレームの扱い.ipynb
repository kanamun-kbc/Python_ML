```python
import pandas as pd

datalist = [['Alice', 25, 'Engineer'],    # データを２次元のリストで定義
        ['Bob', 30, 'Manager'],
        ['Charlie', 35, 'Scientist']]

retsumei = ['Name', 'Age', 'Job']           # 列名のリストを定義

# データフレームオブジェクトを作成
df1 = pd.DataFrame(datalist, columns = retsumei)

df1.index = ['A','B','C']                   # 行名のリストを後から追加

df1.columns = ['namae','toshi','shigoto']   # 列名のリストを変更

print(df1)
```

```python
import pandas as pd

data = {                           # データを辞書で作成
    'Region': ['Asia', 'Europe', 'Asia', 'Africa', 'Europe'],
    'Population': [1000000, 500000, 1500000, 800000, 600000]
}
df1 = pd.DataFrame(data)           # データフレームオブジェクトの作成
df2 = df1.sort_values('Population',ascending=False)
print(df2)
```

```python
import pandas as pd

# 商品データの作成
data = {
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Books'],
    'Price': [1000, 1500, 2000, 2500, 3000]
}
df1 = pd.DataFrame(data)
print(df1)

# 商品カテゴリごとの平均価格を計算
groupA = df1.groupby('Category')
mean_price = groupA.mean()
print(mean_price)
```

```python
import pandas as pd

# 地域データの作成
data = {
    'Region': ['Asia', 'Europe', 'Asia', 'Africa', 'Europe'],
    'Population': [1000000, 500000, 1500000, 800000, 600000]
}
df1 = pd.DataFrame(data)
print(df1)

# 地域ごとの人口総数を計算
groupR = df1.groupby('Region')
total_population = groupR.sum()
print(total_population)
```
