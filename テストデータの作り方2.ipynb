(1)の回答  2次元散布図の作成

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import japanize_matplotlib

X, y = make_blobs(n_samples=300, centers=4, n_features=2,
                  random_state=42, cluster_std=1.0)
# print(X)
# print(y)

plt.scatter(X[:,0:1], X[:,1:2], c = y, alpha=0.9)

plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

(2)の回答  平行移動と色分け

```python
# 平行移動分の加算
X_moved = X + 3.5
# 色指定、赤、青、緑、橙
colorpallet = ["r","b","g","y"]
color = np.array([(colorpallet[i]) for i in y])
# それぞれをプロット
plt.scatter(X[:,0:1], X[:,1:2], c = color, alpha=0.9)
plt.scatter(X_moved[:,0:1], X_moved[:,1:2], c = color, alpha=0.5)
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
