```python
# １つの画像を NumPy配列 にして返す関数
def img2array(img_path):
    img_c1 = Image.open(img_path).convert("L")
#    plt.imshow(img_c1, cmap="gray")
#    plt.show()
    img_c2 = img_c1.resize((8,8),Image.LANCZOS)
#    plt.imshow(img_c2, cmap="gray")
#    plt.show()
    img_arr1 = np.asarray(img_c2, dtype=float)
#    print(img_arr1)
    img_arr2 = 16 - np.floor(17 * img_arr1/256)
#    print(img_arr2)
#    plt.imshow(img_arr2, cmap="gray")
#    plt.show()
    img_arr3 = img_arr2.reshape(1,-1)
#   print(img_arr3.shape)
#   print(img_arr3)
    img_arr4 = np.array([[]])
    if "maru" in img_path:
        img_label = 0
    elif "sankaku" in img_path:
        img_label = 1
    elif "shikaku" in img_path:
        img_label = 2
    elif "batsu" in img_path:
        img_label = 3
    else:
        img_label = -1

    img_arr4 = np.append(img_arr3, [[img_label]], axis=1)
#   print(img_arr4)
    return img_arr4
#   return img_arr3
```

```python
# 複数の画像ファイル名リストからCSVファイルを作る関数
def imgs2csv(img_list, csv_name):
    csv_array = np.empty((0,65),float)
    for img_path in img_list:
        # print(img2array(img_path))
        # print(img_path)
        #csv_array = np.append(img2array(img_path), csv_array, axis=0)   ←　痛恨のミス！！！！！
        csv_array = np.append(csv_array, img2array(img_path), axis=0)

    h_str = ",".join(["x"+str(i) for i in range(1,66)])
    np.savetxt(csv_name, csv_array, delimiter=",", comments="", header=h_str)
    return csv_array
```

```python
# folder2csv()関数の引数について
# 第１引数：参照するフォルダのパス
# 第２引数：出力するCSVファイルの名
# 第３引数：変換したい画像の拡張子（.で始まる）のリスト

def folder2csv(dir_path, csv_name, ex_list):
    file_list = glob.glob(dir_path + '*.*')    # dir_path + '*.*' と打つと、拡張子があるファイルのみがリスト化される。
    img_list = []    # 空リストを作る（あとでファイル名を追加するつもり）

    for file_name in file_list:
        name, ext = os.path.splitext(file_name)
        # print(ext)
        if ext in ex_list:    # ・・・・①
        #if ext == ".png"  or  ext == ".jpg": でも良いが、
        #①のように folder2csv()関数の第３引数に存在するか否かを条件にする方がラク！
            img_list.append(file_name)
    print("あなたが指定したフォルダ内の画像リストは")
    print(img_list)
    img_array = imgs2csv(img_list, csv_name)
    return img_array
```

# (2)データを分けて学習、正解率を求める

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# CSVファイルの読み込み
df = pd.read_csv("enshu15_1.csv")

# 特徴変数と目的変数の分割
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# 学習データとテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# ロジスティック回帰モデルの作成
model = LogisticRegression(max_iter=1000)

# モデルの学習
model.fit(X_train, y_train)

# テストデータの予測
y_pred = model.predict(X_test)

# 正解率の計算
accuracy = accuracy_score(y_test, y_pred)

print(f"テストデータの正解率は {accuracy * 100:.2f} %")
```
