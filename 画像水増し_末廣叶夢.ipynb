# 画像水増しコード

内容はシンプル。

用意された元画像集(ここでは初期に用意された205枚?)に対して

角度[40, 80, 120, 160, 200, 240, 280, 320, 360]度の回転を指定すると、

1枚の画像情報から9枚の画像情報に増やしたことになる。

この場合は205 × 9 = 1845枚


つまり、anglesの内容をもっと増やしてやれば、回転画像の複製については増やすことが可能。

やろうと思えば、ズームや平行移動などのパラメータもあるので、

変数化して色々試してもいいかも。

規則的な角度にしている理由は、ランダムにしてしまうと、

同じ画像が生成されてしまう可能性があるため、それを避けたかったから。

```python
from PIL import Image
import numpy as np
from keras.preprocessing import image
from keras.preprocessing.image import img_to_array, array_to_img
import os
# 画像のパス(初期に用意された画像があるフォルダ)(増やしたい画像のこと)
image_folder_path = '/content/drive/My Drive/Colab Notebooks/png_sharesam1128/'
# 保存パス(ここで複製した画像を保存するフォルダ指定)(後でモデル学習で使う)
image_folder_path2 = '/content/drive/My Drive/Colab Notebooks/hukusei_gazou5/'
# 画像フォルダ内の全てのファイルをリストアップ
image_files = os.listdir(image_folder_path)
# 図形の回転角度
angles = [20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 320, 340, 360]
# 各画像に対して処理を行う
for image_file in image_files:
    # 画像の読み込み
    img = Image.open(os.path.join(image_folder_path, image_file))
    # 各角度に対して処理を行う
    for angle in angles:
        # 画像をnumpy arrayに変換
        img_array = img_to_array(img)
        # 回転
        rotated_img_array = image.apply_affine_transform(
            img_array,
            theta=angle,    # 回転角度
            tx=0,           # 水平方向の移動
            ty=0,           # 垂直方向の移動
            shear=0,        # シアー変換
            zx=1,           # 水平方向のズーム
            zy=1,           # 垂直方向のズーム
            row_axis=0,     # 行の軸
            col_axis=1,     # 列の軸
            channel_axis=2, # カラーチャンネルの軸
            fill_mode='constant',  # 回転後の塗りつぶしモード
            cval=255        # 塗りつぶしの値（白色）
        )
        # 回転した画像をImageオブジェクトに変換
        rotated_img = array_to_img(rotated_img_array)
        # 画像の保存
        save_path = os.path.join(image_folder_path2, f"{os.path.splitext(image_file)[0]}_rotated_image_{angle}.png")
        rotated_img.save(save_path)
        print(f'Image {image_file} rotated at {angle} degrees saved at {save_path}')
```
